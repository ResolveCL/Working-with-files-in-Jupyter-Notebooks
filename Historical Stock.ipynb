```python
import yfinance as yf
import pandas as pd

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import yfinance as yf
import pandas as pd

gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
import requests
import pandas as pd

api_key = "MT3R6F1W595H11I0"
symbol = "GME"

url = f"https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={symbol}&apikey={api_key}"
response = requests.get(url)
data = response.json()

quarterly_reports = data["quarterlyReports"]
gme_revenue = pd.DataFrame(quarterly_reports)[["fiscalDateEnding", "totalRevenue"]]
gme_revenue.columns = ["Date", "Revenue"]

gme_revenue = gme_revenue[gme_revenue["Revenue"] != "None"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].astype(float)
gme_revenue.tail()
```

```python
import plotly.graph_objects as go

def make_graph(stock_data, stock, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Open'], mode='lines', name=stock))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Stock Price (USD)')
    fig.show()
make_graph(tesla_data, 'TSLA', 'Tesla Stock Opening Prices Over Time')
```

```python
make_graph(gme_data, 'GME', 'GameStop Stock Opening Prices Over Time')
```
